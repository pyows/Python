# 데이터 시각화 (Seaborn)
* 홈페이지 : https://seaborn.pydata.org/
* matplotlib 기반의 python 데이터 시각화 라이브러리
* 설치 : conda install seaborn 또는 pip install seaborn
* 표현 가능한 그래프
> 관계형, 카테고리, 분포, 회귀분석, 멀티플롯, 스타일, 색상 등

```python
# 라이브러리 임포트

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#sns.set(style='darkgrid')

%matplotlib inline
```

```python
# Seaborn 에서 제공한는 데이터셋 

tips = sns.load_dataset('tips')
tips.head()
```

> total_bill : 전체 가격<br>
tip : 팁 가격<br>
sex : 팁을 준 사람의 성별<br>
smoker : 흡연 여부 (yes/no : 범주형 데이터)<br>
day : 요일<br>
time : 방문시간<br>
size : 방문자수

```python
sns.relplot(data=tips, x='total_bill', y='tip')
```

```python
sns.set(style='darkgrid')   # 그래프 스타일을 darkgrid로 적용(회색의 격자무늬)
```

```python
sns.relplot(data=tips, x='total_bill', y='tip')
```

> 비용이 클 수록, 팁도 크다 (우상향)

### 색상을 적용하여, 흡연자와 비흡연자를 표시

```python
# hue : 색상 적용 (범주형 데이터)

sns.relplot(data=tips, x='total_bill', y='tip', hue='smoker' )   # smoker는 범주형 데이터
```

```python
# hue : 색상 적용 (수치형 데이터)

sns.relplot(data=tips, x='total_bill', y='tip', hue='size' )   # size는 수치형 데이터
```

### Marker의 스타일 적용

```python
# style : marker의 스타일 적용 

sns.relplot(data=tips, x='total_bill', y='tip', style='smoker', hue='size')   # 
```

> style : 마커의 모양을 다르게 적용 (O/X)<br>
size : 색상을 다르게 적용(값에 따라 농도를 다르게)

### 색상 선택하기 (palette)
>https://seaborn.pydata.org/tutorial/color_palettes.html<br>
https://seaborn.pydata.org/generated/seaborn.color_palette.html

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='size', palette='Blues') 
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='size', palette='ch:2.5,-.2,dark=.3') 
```

### Marker의 사이즈 설정하기 (size)

```python
sns.relplot(data=tips, x='total_bill', y='tip', size='size')  # 값에 따라 마커의 크기 변경
```
