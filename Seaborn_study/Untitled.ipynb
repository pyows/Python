# 선형그래프 그리기

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid')

%matplotlib inline
```

```python
df = pd.DataFrame(dict(time=np.arange(500), value=np.random.randn(500).cumsum()))
df.head()
```

### 선형그래프 그리기 (kind='line)

```python
sns.relplot(data=df, x='time', y='value', kind='line')
```

### 새로운 그래프 그리기

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
sns.relplot(data=fmri , x='timepoint', y='signal', kind='line')
```

```python
sns.relplot(data=fmri , x='timepoint', y='signal', kind='line', ci=None)
```

```python
sns.relplot(data=fmri , x='timepoint', y='signal', kind='line', ci='sd')
```

```python
sns.relplot(data=fmri , x='timepoint', y='signal', kind='line', hue='region', style='event')
```

# 범주형 데이터를 사용한 그리기 (catplot)

### cat plot : 범주형 데이터에 대한 그리기
### relplot : 관계형 데이터에 대한 그리기

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.catplot(data=tips, x='day', y='total_bill')     # 요일별 요금 정산금액
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='swarm')  # 마커가 겹치지 않도록 표시(kind='swarm')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='box')   
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='violin')   # 바이올린 그래프 (좌/우로 나누어 볼때, 양쪽의 비율을 확인할 때)
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='boxen')   # kind='boxen'
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='count')    # 셀 수 없는 데이터이기 때문에 에러발생
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='bar')  # 검정색 : 편차를 나타냄
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='point')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='swarm', hue='sex') 
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='violin', hue='sex') 
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='violin', inner='stick', hue='sex')   # inner='stick'
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='violin', inner='stick', split=True, hue='sex') # split
```
